# 🏢 Interprétation des statistiques descriptives (df.describe)

## 1. Vue d’ensemble
Le dataset contient **10 000 biens immobiliers publics** avec 13 colonnes :
- Identifiants (`ue`, `id`)
- Localisation (`region`, `dept`, `adresse`, `code_postal`, `ville`, `pays`)
- Usage (`type`, `fonction`, `designation_batiment_terrain`)
- Ministère occupant (`ministere`)
- Date de l’inventaire (`date_inventaire`)

---

## 2. Analyse par colonnes

### Identifiants
- **id** : 10 000 valeurs mais seulement 9 041 uniques → certains identifiants apparaissent plusieurs fois (doublons à traiter).
- **ue** : 9 703 valeurs renseignées, 5 110 uniques. Valeurs manquantes (≈3 %) car remplacé par l’`id` après 2015.

### Localisation
- **region** : 25 valeurs distinctes, cohérent avec les anciennes régions françaises. Valeur la plus fréquente : *Île-de-France* (1171 biens).
- **dept** : 110 départements distincts. 2 valeurs manquantes.
- **ville** : 6 972 valeurs renseignées (≈70 %). 3 775 villes distinctes. Valeur la plus fréquente : *La Faute-sur-Mer* (54 occurrences).
- **code_postal** : 6 985 valeurs renseignées (≈70 %), 2 310 distincts.
- **adresse** : seulement 6 712 adresses renseignées (≈67 %), 5 697 distinctes.
- **pays** : quasi exclusivement *France* (98,2 %).

### Usage
- **type** : 18 catégories différentes. Valeur la plus fréquente : *Espace naturel* (3 887 biens).
- **fonction** : 107 usages détaillés. Valeur la plus fréquente : *biens du ministère de l’Intérieur* (2 989 occurrences).
- **designation_batiment_terrain** : 5 515 désignations distinctes, mais redondance forte avec *biens du ministère de l’Intérieur* (2 989 occurrences).

### Ministère
- **ministere** : 46 ministères distincts. Le plus représenté : *Écologie - Développement durable - Énergie ...* (2 518 biens).

### Date d’inventaire
- **date_inventaire** : seulement 6 valeurs distinctes (ex. 2013-05, 2014-03, 2015-10…). Cela correspond à des campagnes d’inventaire ponctuelles.

---

## 3. Points clés pour le nettoyage
- 🔍 **Doublons** : vérifier `id`, qui devrait être unique.
- ⚠️ **Valeurs manquantes** : fortes proportions sur `adresse`, `code_postal`, `ville`.
- 🧹 **Normalisation** :
  - Harmoniser la casse (`ville`, `region`, `ministere`).
  - Nettoyer les désignations redondantes.
- 📊 **Colonnes peu informatives** :
  - `pays` (quasi toujours *France*).
  - `date_inventaire` (peu de valeurs distinctes).

```python
# Affichage des valeurs distinctes pour les colonnes d'intérêt
for col_name in ["type", "fonction", "region", "dept", "ministere"]:
    valeurs = df[col_name].dropna().unique()
    valeurs.sort()
    print(f"\n{'='*50}")
    print(f"  {col_name.upper()} ({len(valeurs)} valeurs)")
    print(f"{'='*50}")
    for v in valeurs:
        print(v)
```
